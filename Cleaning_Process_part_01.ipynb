```python
df.loc[df['date'] == 'March 10, 2024', 'date'] = '10-03-2024'
df.loc[df['date'] == '2024/04/05', 'date'] = '05-04-2024'
df.loc[df['date'] == '2024-01-01', 'date'] = '01-01-2024'
df['date'].unique()
```

```python
# check the format and change the date na fillna ()--method used
df['date']=df['date'].fillna(method='bfill')
```

```python
# Convert from string (object) to datetime, specifying dayfirst=True
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y', errors='coerce')
```

```python
# atta apalykade ase 11 rows ahet tyanchi value 'missing' ahe tya missing la null ne replace karayache.
# nantar column la string to int karayche 
# mg null value la mean ne fill karayache
df['Column_name'] = df['Column_name'].replace('missing', np.nan)
```

```python
# Step 1: Convert to numeric (this will turn non-numeric like NaN or 'missing' into np.nan)
df['clicks'] = pd.to_numeric(df['clicks'], errors='coerce')
df['clicks'].dtype
```
